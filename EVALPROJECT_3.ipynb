### Feature Engineering and handling outliers

```python
df.head()
```

```python
sex=pd.get_dummies(df["sex"],drop_first=True)
sex.head()
```

```python
df=pd.concat([df,sex],axis=1)
df.head()
```

```python
df.drop(["sex"],axis=1,inplace=True)
df.head()
```

### now we handle the outliers

```python
from scipy import stats
# outliers=df.iloc[:,1:]
z=np.abs(stats.zscore(df))
print(z)
```

```python
threshold=3
print(np.where(z>3))
```

```python
print("the shape of the dataframe before removing outliers",df.shape)
newdf = df[(z < 3).all(axis=1)]
print("the shape of the dataframe after removing outliers",newdf.shape)
# # newdf = newdf[(z < 3).all(axis=1)]
```

```python
sns.pairplot(data=newdf)
```

### Train test split 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor,ExtraTreesRegressor
```

```python
newdf.head()
```

```python
x=newdf.loc[:,["length","diameter","height","whole_weight","shucked_weight","viscera_weight","shell_weight","I","M"]]
x.head()

y=newdf.loc[:,["rings"]]
y.head()
```

```python
count=0
numbers=range(101)
model=[LinearRegression(),Lasso(),Ridge()]
for m in model:
    for i in numbers:
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
        m.fit(x_train,y_train)
        score=m.score(x_train,y_train)
        y_pred=m.predict(x_test)
        print("the score for the corresponding random state",i,"is",score)
        
        if score>count:
            count=score
            final_i=i
    print("The model used is :", m)

    print("\n")
print("the max score for random state",final_i,"is",count)
```

```python
count=0
numbers=range(101)
model=[RandomForestRegressor(n_estimators=200),AdaBoostRegressor(n_estimators=200),GradientBoostingRegressor(n_estimators=200),ExtraTreesRegressor(n_estimators=200)]
for m in model:
    for i in numbers:
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
        m.fit(x_train,y_train)
        score=m.score(x_train,y_train)
        y_pred=m.predict(x_test)
        print("the score for the corresponding random state",i,"is",score)
        
        if score>count:
            count=score
            final_i=i
    print("The model used is :", m)

    print("\n")
print("the max score for random state",final_i,"is",count)
```

### selecting the best algorithm

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=0)
rfr=RandomForestRegressor(n_estimators=200)        
rfr.fit(x_train,y_train)
score=cross_val_score(rfr,x,y,cv=10,scoring="neg_mean_squared_error")
y_pred=cross_val_predict(rfr,x,y,cv=10)

print(score)
# print(score.mean())
print("\n")
print(y_pred)
nscore=rfr.score(x_test,y_test)
print(nscore)

# plt.plot(y,y_pred)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=0)
etr=RandomForestRegressor(n_estimators=200)
etr.fit(x_train,y_train)
score=etr.score(x_train,y_train)
y_pred=etr.predict(x_test)
print(score)
```

```python
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

CONCLUSION:

1)We use cross validation, but as we can see that it did not made much of a differnce in the score hence i can say that the scores are pretty well balanced and we dont have to use cross validation.

2)From the above for loops randomforest and extratrees have performed the best, but as extratrees will always overfit we will go ahead and use the randomforest model further . RandomForest has outperformed all the other models

### Finalising the Model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=0)
lr=RandomForestRegressor(n_estimators=200,max_depth=125,oob_score=False)
lr.fit(x_train,y_train)
score=lr.score(x_train,y_train)
y_pred=lr.predict(x_test)
print(score)
```

```python
print("r2_score : ",r2_score(y_test,y_pred))
print("RMSE : ",np.sqrt(mean_squared_error(y_test,y_pred)))
#we have a low r2 score but our errors are also quite low .
```
